```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
PATIENT_N = 100000
sicknes_ratio = 0.4
patient_sick = sicknes_ratio * PATIENT_N
patient_healthy = PATIENT_N - patient_sick
positive_sick = patient_sick * 0.95
positive_healty = patient_healthy * 0.55
print(positive_sick / (positive_sick + positive_healty))
```

```python
#b
PATIENT_N = 100000
sicknes_ratio = 0.1
patient_sick = sicknes_ratio * PATIENT_N
patient_healthy = PATIENT_N - patient_sick
positive_sick = patient_sick * 0.95
positive_healty = patient_healthy * 0.55
print(positive_sick / (positive_sick + positive_healty))
```

```python
x = np.linspace(0, 1, 100)
y = 0.95 * x / (0.95 * x + 0.55 * (1-x))

plt.plot(x, y, 'r', label='posterior')
plt.plot(x, x, 'black', label='y=x', linestyle='--')
plt.legend()
plt.show()
```
