```python
import numpy as np
import pandas as pd
```

```python
player_df = pd.read_csv('../kaggle/players.csv')
player_df = player_df[player_df['last_season'] >= 2023]
player_df = player_df[player_df['current_club_domestic_competition_id'].isin(['GB1', 'ES1', 'IT1', 'L1'])] # only premier, la liga serie a and bundesliga
player_df = player_df[player_df['position'].isin(['Midfield', 'Attack'])]
player_df
```

```python
appear_df = pd.read_csv('../kaggle/appearances.csv')
appear_df = appear_df[appear_df['player_id'].isin(player_df['player_id'])]
appear_df = appear_df.groupby(['player_id']).agg({'goals': 'sum', 'assists': 'sum', 'minutes_played': 'sum'})
df = pd.merge(appear_df, player_df, left_on='player_id', right_on='player_id', how='left')
df['log_minutes_played'] = np.log(df.minutes_played)
df['log_highest_value'] = np.log(df.highest_market_value_in_eur)
df
```

```python
df.to_csv('player_stats.csv')
```
